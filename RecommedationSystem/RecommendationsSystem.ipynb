```python
import langchain
import openai
import numpy as np
import chromadb
```

```python
from langchain.embeddings import FakeEmbeddings

fake_embeddings = FakeEmbeddings(size=1536)

fake_embeddings_list = []

for quote in life_and_programming_sentences:
    fake_embeddings_list.append(fake_embeddings.embed_query(quote))
```

```python
fake_embeddings_list = np.array(fake_embeddings_list).astype("float32")
```

```python
# Vector database
from chromadb.config import Settings

client = client = chromadb.Client(Settings(chroma_db_impl="duckdb+parquet",
                                    persist_directory="db/"
                                ))
collection = client.create_collection(name="Comments", embedding_function=fake_embeddings)
collection.add(documents=life_and_programming_sentences,
               ids=[f"id{i}" for i in range(len(life_and_programming_sentences))]
              )
```

```python
# Semantic Analysis
query_results = collection.query(query_texts=["I love programming"],
                                 n_results=10,
                                )
query_results
```
